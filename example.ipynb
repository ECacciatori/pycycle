## PyCycle WIP

```python
import pycycle 

import pandas as pd
import plotly.graph_objs as go
import plotly.figure_factory as FF

from plotly import tools
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)
```

### Load an activity from Strava

```python
strava_access_token = 'XXXXXXX'

df = pycycle.from_strava(1138745677, strava_access_token)
```

```python
df.head()
```

### Compute additional columns

```python
df['Power_Smooth'] = df['Power'].rolling(window=3).mean()
df['Cadence_Smooth'] = df['Cadence'].rolling(window=3).mean()
```

### Plot basic data

```python
power_trace = go.Scatter(x=df.index, y=df['Power_Smooth'], name='Power (W)')
cadence_trace = go.Scatter(x=df.index, y=df['Cadence_Smooth'], name='Cadence (RPM)')
distance_trace = go.Scatter(x=df.index, y=df['Velocity'], name='Velocity (m/s)')

fig = tools.make_subplots(rows=3, cols=1, shared_xaxes=True, print_grid=False)
fig.append_trace(power_trace, 1, 1)
fig.append_trace(cadence_trace, 2, 1)
fig.append_trace(distance_trace, 3, 1)
fig['layout']['title'] = 'Power/Cadence/Velocity'
iplot(fig)
```

```python
hist_data = [df['Power']]
group_labels = ['Power']
fig = FF.create_distplot(hist_data, group_labels, curve_type='normal', bin_size=20)
fig['layout']['title'] = 'Power Distribution (20W)'
iplot(fig)
```
